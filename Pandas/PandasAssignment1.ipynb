Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

DataFrame is a two-dimensional and stores heterogeneous tabular data structure with labeled axes(rows and columns).

In pandas,Series is a one-dimensional labeled array whereas DataFrame is a two-dimensional table where each column is a pandas Series.

```python
lst=[23,29,48,29,59,17,40,60,48,40]
series=pd.Series(lst)
series
```

```python
data={"Name":["Alice","Bob","Claire"],"Age":[25,30,27],"Gender":["Female","Male","Female"]}
df2=pd.DataFrame(data)
df2
```
